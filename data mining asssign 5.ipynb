```python
def sigmoid(X,w):
    return 1/(1+np.exp(-np.inner(w,X)))
def regularised_logistic(X,y,a,l,tol=1):
    plot_data = []
    w = 10*np.random.random(len(X[0]))
    I = np.zeros((len(w),len(w)))
    I[1:,1:] = np.eye(len(w)-1)
    Xtrain,Xvalid,Ytrain,Yvalid = train_test_split(X,y,test_size=0.2,random_state=0)
    J = 20
    while True:
        Jold = J
        for i in range(len(Xtrain)):
            f = sigmoid(Xtrain[i],w)
            w = w + a*(Ytrain[i]-f)*Xtrain[i] - l*(w@I)
        J = 0
        for j in range(len(Xvalid)):
            f = sigmoid(Xvalid[j],w)
            J += (-Yvalid[j]*np.log(f)-(1-Yvalid[j])*np.log(1-f) + 0.5*l*np.linalg.norm(w@I))
        plot_data.append(J)
        if J <= tol:
            break
        elif Jold == J:
            break
        print(J)
    return w,plot_data
```

```python
def add_bias(data):
    data.insert(0,0,1,allow_duplicates=True)
    data.columns = pd.RangeIndex(len(data.columns))
    return data
```

```python
def poly_fit(Data,order):
    X=Data.copy()
    for i in range(2,order+1):
        X[i]=X[i-1]*X[1]
    return X
```

```python
def poly_reg(X,Y):
    pinv = np.linalg.pinv(X)
    W= pinv.dot(Y)
    return W
 
```

```python
def read_data(data):
    cwd=os.getcwd()
    path=os.path.join(cwd,'assign_5', data)
    df=pd.read_csv(path,delimiter=',',header=None,na_values='?')
    return df
```

```python
def train_test(data):
    trn,tst = train_test_split(data, test_size = 0.3,shuffle=True)
    return trn,tst
```

```python
def min_max_norm(data,data_min,data_max):
    for column in data.columns:
        data[column] = (data[column] - data_min[column])/ (data_max[column]-data_min[column])    
    return data
```

```python
def knn_classifier(X_trn,X_tst,k,y_trn):
    ind=X_trn.index
    y_p=[]
    for i in X_tst.values:
        dist=[]
        for j in X_trn.values:
            d=LA.norm(i-j)
            dist.append(d)
        df=pd.Series(dist,index=ind)
        df.sort_values(inplace=True)
        p=y_trn[df.index[:k]].mode()
        y_p.append(int(p))
    return y_p
```

```python
# Reading the data
df=read_data('breast-cancer-wisconsin.data')
df.drop(columns=0,inplace=True) #Dropping the ID column
df.dropna(inplace=True) #Dropping rows with Nan values

# Splitting to X and y parts 
X=df.iloc[:,:-1]
y=df.iloc[:,-1]

#K-fold cross validation
folds = 5
kf = KFold(n_splits=folds, random_state=None) 
Acc=[] 
for train_index , test_index in kf.split(X):
    X_trn , X_tst = X.iloc[train_index,:],X.iloc[test_index,:]
    y_trn , y_tst = y.iloc[train_index] , y.iloc[test_index]


    # Normalising the data using min_max method
    data_min,data_max=X_trn.min(),X_trn.max()
    X_trn=min_max_norm(X_trn.copy(),data_min,data_max)
    X_tst=min_max_norm(X_tst.copy(),data_min,data_max)

    k=3

    y_p=knn_classifier(X_trn,X_tst,k,y_trn)
    match=np.equal(y_p,y_tst).sum()
    Acc.append(match*100/X_tst.shape[0])

# Result for k-fold Method
print("\nAccuracy of folds is",Acc)

# Hold-out method
k=[3,5,7,11]
Acc=[] 
for k in k:
    y_p=knn_classifier(X_trn,X_tst,k,y_trn)
    match=np.equal(y_p,y_tst).sum()
    Acc.append(match*100/X_tst.shape[0])
    
# Result for Holdout Method
print("\nAccuracy for different k's is",Acc)
    
```
